```python
## Script Index: S1
## Author: Liu Yue
## Date: 30/09/2016
## Python Version: Anaconda 4.2.0, python 3.5

""" 
    Data Scraping:
    ZOL is a Chinese cite for phone comparison/blogging/selling
    On the phone-list page, we are able to get link to phone-homepage and specifications
    This script gets the link to phone-list pages of various brands in TalkingData Kaggle Challenge dataset
"""

from bs4 import BeautifulSoup
import urllib.request
import csv
import re


def bsObjFormation(url):
    """ method for creating bs4 object in general """
    user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
    headers={'User-Agent':user_agent,}
    try:
        request=urllib.request.Request(url,None,headers) #The assembled request # may have TimeoutError
        htmlfile=urllib.request.urlopen(request)
        bsObj = BeautifulSoup(htmlfile.read(), "html.parser", )
        return bsObj
    except Exception:
        pass

## tk_brands from the output of mysql
with open('S0_tk_phone_brand.csv', newline='',encoding='utf-8') as f:
    file = csv.reader(f)
    tk_phone_brand = list(file)
tk_brands = sum(tk_phone_brand, [])
tk_brands = [s.upper() for s in tk_brands]

## The zol page with all zol brands
zol_brands = []
bsObj = bsObjFormation("http://detail.zol.com.cn/category/57.html")
if bsObj is not None:
    # Part 1 of the page: Brand name with image displayed on the page
    normal = bsObj.findAll("ul",class_="manu normal")[0]
    for i in range(0,len(normal)):
        line = normal.findAll("li")[i].a
        link = line["href"]
        brand_name = line.img["alt"].encode('cp1252').decode('gbk') 
        zol_brands.append([brand_name,link]) 

    ## Part 2 of the page: Pure text for brand names on the page
    list2 = bsObj.findAll("ul",class_="manu txt")
    for j in range(0, len(list2)):
        sublist = list2[j]
        for i in range(0,len(sublist)):
            line = sublist.findAll("li")[i].a
            link = line["href"]
            if link!='javascript:void(0)':
                brand_name = line["title"].encode('cp1252').decode('gbk') 
                zol_brands.append([brand_name,link]) 

## For all tk_brands found in zol_brands, get name and link of all brands
brands_found = []

for i in range(0, len(zol_brands)):
    zol_brand = zol_brands[i][0].upper()
    for tk_brand in tk_brands:
        if tk_brand in zol_brand: # contained in string        
            line = [tk_brand]
            line.extend(zol_brands[i])
            brands_found.append(line)

## Write the data to CSV files: desired tk_brand_name, brand_name_displayed, zol_link
with open('S1_zol phone brand link.csv','w',newline='',encoding='utf-8-sig') as f:
    a = csv.writer(f)
    a.writerows(brands_found)
```

```python
## How to get back the list

## Read tk_brands_found: tk_brand_name, brand_name_displayed, zol_link
with open('S1_zol phone brand link.csv', newline='',encoding='utf-8') as f:
    file = csv.reader(f)
    lst = list(file)
lst[0][0] = lst[0][0].replace(u'\ufeff', '') # Remove Byte Order Mark character at start of file
brands_found == lst
```
